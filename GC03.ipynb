```python
#Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

graan = pd.read_csv('data/GRAIN.csv',encoding='ISO-8859-1"')

graan.head()
```

```python
#removing whitespace
graan['Status of deal'] = graan['Status of deal'].str.strip()
graan['Landgrabbed'] = graan['Landgrabbed'].str.strip()
graan['Landgrabber'] = graan['Landgrabber'].str.strip()
graan['Base'] = graan['Base'].str.strip()
graan['Sector'] = graan['Sector'].str.strip()
graan['Production'] = graan['Production'].str.strip()
graan['Summary'] = graan['Summary'].str.strip()
graan = graan.convert_objects(convert_numeric=True)

graan.info()
```

```python

#Autofilter
grouped = graan.groupby(by='Status of deal')
grouped.count()
```

```python
#Data validation
index =0;
currentIndex =''
for x in list(graan.columns.values):
    index=0
    for a in graan[x]:
        try:
            graan.loc[index,x] = a.strip();
        except:
            graan.loc[index,x] = a;
        index = index +1;


grouped = graan.groupby(by='Status of deal')
grouped.count()
```

```python
#Disposing of NAN values...
grouped2 = graan.replace({np.nan:'missing'},regex=True)
grouped2.tail()
```
