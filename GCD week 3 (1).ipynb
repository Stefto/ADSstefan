2	Data Validation

```python
# In dit geval zijn de kolommen al een bepaalde waarde toegewezen wat je op de volgende manier kan zien:

df.dtypes
```

```python
# Maar je kan ze expliciet nog een keer een type toewijzen, bijvoorbeeld string en float:

df["Landgrabbed"] = df["Landgrabbed"].astype('str')
df["Hectares"] = df["Hectares"].astype('float64')

df.dtypes
```

3 Defined data ranges

```python
# Bekijk alle kolommen
df.columns.values
```

```python
# Groepeer de waardes van een kolom en kijk wat er allemaal in staat
statusofdeals = df.groupby('Status of deal')
statusofdeals.count()
```

```python
# Verwijder de witruimtes van de waardes

df['Status of deal'] = df['Status of deal'].str.strip()
statusofdeals = df.groupby('Status of deal')
statusofdeals.count()
```

```python
# Nog een voorbeeld van dubbele waardes met witruimtes

base = df.groupby('Base')
base.count()
```

```python
# Daarom is het belangrijk om dit ook voor de rest van de kolommen te doen.

df['Landgrabbed'] = df['Landgrabbed'].str.strip()
df['Landgrabber'] = df['Landgrabber'].str.strip()
df['Base'] = df['Base'].str.strip()
df['Sector'] = df['Sector'].str.strip()

base = df.groupby('Base')
base.count()
```

```python
# Je kan ook nog ervoor kiezen om bij elke kolom de NaN's te verwijderen:

df2 = df.dropna()
df2.head()

# Je ziet zo wel dat er gelijk een heleboel regels missen.
```

4 Find and Replace

```python
# 2 voorbeelden van het zoeken naar een naam en die vervangen door iets anders.

# Oude dataset:
df.head()
```

```python
df.ix[df.Landgrabbed.isin(['Algeria']), 'Landgrabbed'] = 'DZ'  
df.ix[df.Landgrabbed.isin(['Angola']), 'Landgrabbed'] = 'AO'
  
# Nieuwe dataset:
df.head()
```

6 Pivot tables

```python
# Ik heb 2 voorbeelden van hoe je kolommen kan samenvatten:
# Samenvatten van de kolom "Landgrabbed"

pd.pivot_table(df,index=["Landgrabbed"])
```

```python
# Samenvatten van een kolom binnen een andere kolom. Deze gaat dus eerst samenvatten op "Sector" en daarin weer samenvatten op "Landgrabber" 

pd.pivot_table(df,index=["Sector", "Landgrabber"])
```

7 Regular Expressions

```python
# Hij zoekt voor alle woorden die in die kolom staan.

import re
pattern = r'[a-z]+'
result = df['Projected investment'].str.findall(pattern, re.IGNORECASE)
result
```

```python
# Of je kan zoeken op een specifiek woord als er bijvoorbeeld veel tekst in een kolom staat zoals in "Summary":

pattern = r'company'
result = df['Summary'].str.findall(pattern, re.IGNORECASE)
result
```

8 Text-to-Columns

```python
# Ik ga de "Sector" kolom splitten zodat elke sector op een aparte regel komt.

s = df['Sector'].str.split(', ').apply(pd.Series, 1).stack() # split elke kolom op de komma's

s.index = s.index.droplevel(-1) # Zo komt elke nieuwe regel op dezelfde index als de originele regel te staan

s.name = 'Sector' # Komt terug als een nieuwe kolom die je dan weer Sector noemt

del df['Sector'] # Oude Sector kolom verwijderen.

df.join(s)
```
