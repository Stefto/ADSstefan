```python
import numpy as np
import pandas as pd
import sklearn as sk
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline     

df = pd.read_csv("pokerhand.csv")
df.head(5) #checking out the data
```

```python
#making the data readable
def suits(value):
    if(value==1):
        return "Hearts"
    if(value==2):
        return "Spades"
    if(value==3):
        return "Diamonds"
    if(value==4):
        return "Clubs"

def FaceCards(value):
    if(value==1):
        return "Ace"
    if(value==11):
        return "Jack"
    if(value==12):
        return "Queen"
    if(value==13):
        return "King"
    return value

def Hand(value):
    if(value==0):
        return "nothing"
    if(value==1):
        return "Pair"
    if(value==2):
        return "two Pairs"
    if(value==3):
        return "Three of a kind"
    if(value==4):
        return "Straight"    
    if(value==5):
        return "Flush"
    if(value==6):
        return "Full house"
    if(value==7):
        return "Four of a kind"
    if(value==8):
        return "Straight flush"
    if(value==9):
        return "Royal straight flush"

#for i in df.index:
#    df.loc[i,"suit1"] = suits(df.loc[i,"suit1"])
#    df.loc[i,"rank1"] = FaceCards(df.loc[i,"rank1"])
#    df.loc[i,"suit2"] = suits(df.loc[i,"suit2"])
#    df.loc[i,"rank2"] = FaceCards(df.loc[i,"rank2"])
#    df.loc[i,"suit3"] = suits(df.loc[i,"suit3"])
#    df.loc[i,"rank3"] = FaceCards(df.loc[i,"rank3"])
#    df.loc[i,"suit4"] = suits(df.loc[i,"suit4"])
#    df.loc[i,"rank4"] = FaceCards(df.loc[i,"rank4"])
#    df.loc[i,"suit5"] = suits(df.loc[i,"suit5"])
#    df.loc[i,"rank5"] = FaceCards(df.loc[i,"rank5"])


#df.to_csv("pokerhandReadable.csv");
#df.head(5)
len(df.index)
```

```python
allSuits = np.concatenate((df["suit1"].tolist() , df["suit2"].tolist() ,df["suit3"].tolist(), df["suit4"].tolist() ,df["suit5"].tolist()), axis=0);
print(len(allSuits))
allCards =np.concatenate((df["rank1"].tolist() , df["rank2"].tolist() ,df["rank3"].tolist(), df["rank4"].tolist() ,df["rank5"].tolist()), axis=0);
print(len(allCards))

cardCounts = { "Hearts Ace": 0, "Hearts 2": 0,"Hearts 3": 0, "Hearts 4": 0,"Hearts 5": 0, "Hearts 6": 0,"Hearts 7": 0, "Hearts 8": 0,"Hearts 9": 0, "Hearts 10": 0, "Hearts Jack": 0, "Hearts Queen": 0,"Hearts King": 0, "Spades Ace": 0, "Spades 2": 0,"Spades 3": 0, "Spades 4": 0,"Spades 5": 0, "Spades 6": 0,"Spades 7": 0, "Spades 8": 0,"Spades 9": 0, "Spades 10": 0,"Spades Jack": 0, "Spades Queen": 0,"Spades King": 0,"Diamonds Ace": 0, "Diamonds 2": 0,"Diamonds 3": 0, "Diamonds 4": 0,"Diamonds 5": 0, "Diamonds 6": 0,"Diamonds 7": 0, "Diamonds 8": 0,"Diamonds 9": 0, "Diamonds 10": 0, "Diamonds Jack": 0, "Diamonds Queen": 0,"Diamonds King": 0,"Clubs Ace": 0, "Clubs 2": 0,"Clubs 3": 0, "Clubs 4": 0,"Clubs 5": 0, "Clubs 6": 0,"Clubs 7": 0, "Clubs 8": 0,"Clubs 9": 0, "Clubs 10": 0, "Clubs Jack": 0, "Clubs Queen": 0,"Clubs King": 0};
count = 0
for i in np.nditer(allSuits):
    suit = suits(i);
    card = FaceCards(allCards[count]);
    cardCounts[suit + " " + str(card)] = cardCounts[suit + " " + str(card)] +1;
    count = count +1;
    
for key in cardCounts:
    print(key + " has " + str(cardCounts[key]) + " cards")
```

```python
#0: Nothing in hand; not a recognized poker hand 
#1: One pair; one pair of equal ranks within five cards 
#2: Two pairs; two pairs of equal ranks within five cards 
#3: Three of a kind; three equal ranks within five cards 
#4: Straight; five cards, sequentially ranked with no gaps 
#5: Flush; five cards with the same suit 
#6: Full house; pair + different rank three of a kind 
#7: Four of a kind; four equal ranks within five cards 
#8: Straight flush; straight + flush 
#9: Royal flush; {Ace, King, Queen, Jack, Ten} + flush 
    
sets = {"nothing": 0, "Pair":0,"two Pairs":0,"Three of a kind":0,"Straight":0,"Flush":0,"Full house":0,"Four of a kind":0,"Straight flush":0,"Royal straight flush":0}
listOfHands = df["hand"].tolist();

for x in listOfHands:
    sets[Hand(x)] = sets[Hand(x)] +1;
    
for key in sets:
    print(key + " has occured " + str(sets[key]) + " times")  
```
